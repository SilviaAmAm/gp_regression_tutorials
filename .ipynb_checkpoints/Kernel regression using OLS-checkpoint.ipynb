```python
import numpy as np
import statsmodels.api as sm
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# np.random.seed(1)
N = 20
x = np.linspace(0, 10, N)
y = np.sin(x) + np.random.normal(0,0.1, size=N)
```

```python
def gauss_kernel(x, knots, w=40):
    # Gaussian kernel using predetermined knots. 
    return np.array([np.exp(-(x-knot)**2/w) for knot in knots]).T
```

```python
K = gauss_kernel(x, knots=x, w=40)
plt.imshow(K)
```

```python
model = sm.OLS(y, K)
result = model.fit()
```

```python
plt.scatter(x, y)
plt.plot(x, result.fittedvalues)
```

```python
result.params
```

```python
Xnew = np.linspace(x.min(), x.max(), 100)
Knew = gauss_kernel(Xnew, x, w=40) 
y_predict = result.predict(Knew)
```

```python
plt.plot(Xnew, y_predict)
plt.scatter(x, y)
```
