```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import pymc3 as pm
sns.set_context('talk')
sns.set_style('whitegrid')
import statsmodels.formula.api as smf
import pandas as pd
```

```python
np.random.seed(1)
N = 20
x = np.random.uniform(0, 10, size=N)
y = np.sin(x) + np.random.normal(0,0.1, size=N)
with sns.plotting_context('talk', font_scale=1):
    plt.plot(x, y, 'o')
    plt.xlabel('$x$')
    plt.ylabel('$f(x)$')
```

 Let's have 5 knots equally spaced across the $X$ axis each with a bandwidth of 2:

```python
n_knots = 5
knots = np.linspace(np.floor(x.min()), np.ceil(x.max()), n_knots)
w = 2
def gauss_kernel(x, knots, w=2):
    # Gaussian kernel using predetermined knots. 
    return np.array([np.exp(-(x-knot)**2/w) for knot in knots])

kernel = gauss_kernel(x, knots, w=2) 
```

For each knot, there is a kernel defined over all the data. i.e. for each knot, there's a value of $K$ for each value of $X$.  Let's plot this: 

```python
with sns.plotting_context('talk'):
    fig, axes = plt.subplots(n_knots, figsize=(5,10), 
                             sharex=True, sharey=True)
    cols = sns.color_palette('colorblind')
    
    for i, ax in enumerate(axes):
        ax.scatter(x, y, color=cols[0], marker='o', label='data')
        ax.scatter(x, kernel[i], color=cols[1], alpha=0.7, marker='x', 
                  label=r'$K(X, q_{%s} )$' % str(i+1))
        ax.vlines(knots[i], -2, 2, color=cols[1], alpha=0.7, 
                  label=r'$q_{%s}$' % str(i+1))
        ax.legend(bbox_to_anchor=(1.1, 1.05))
        ax.set_ylabel('knot %s' % str(i+1))
    ax.set_xlabel('X')
```

## Kernel regression is local
As you can see, the green x's look like Gaussian's (because they are) around the knot center (the vertical line).  

So why do this?  

Well, it allows us to assign a regression coefficient to a small subset of the data points - those that are similar or local to the knot points.  For those data far the first knot, i.e.  $ |x - q_1|^2 \gg w$ so that $K \sim 0$.  This means, the only terms in the regression equation are those near $q_1$.  i.e.: 

$$
\hat{Y} \approx \gamma_1 K(X, q_1) = \gamma_1 \exp\left(-\frac{||X-q_1||^2}{w}\right)
$$

The predicted value of the  response variable approaches $\gamma_1$ as the values of the input variables approach the first knot, $q_1$. For values that fall inbetween $q_1$ and $q_2$, the predicted value of the response variable will be a weighted average of $\gamma_1$ and $\gamma_2$. 

## Conditioning a kernel regression model
Now let's use the Bayesian approach on the kernal regression model.  This proceeds the same way as Bayesian linear regression.  We put priors on the regression coefficients and the noise.  For the gammas we'll use a [Cauchy](https://en.wikipedia.org/wiki/Cauchy_distribution) distrubtion and for the noise we'll use a uniform distribution.

```python
with pm.Model() as kernel_model:
   gamma = pm.Cauchy('gamma', alpha=0, beta=1, shape=n_knots)
   sd = pm.Uniform('sd',0,  10)
   mu =  pm.math.dot(gamma, gauss_kernel(x, knots))
   yl = pm.Normal('yl', mu=mu, sd=sd, observed=y)
   kernel_trace = pm.sample(1000, tune=500)

_ = pm.traceplot(kernel_trace)
```

This seems very well converged: the two posterior distributions from each chain lie on top of one another.

## Plotting the model

Before we plot our predictions let's just review in more detail how a regression model is specified.  This applies to Bayesian and non-Bayesian models.  

For each type of regression model we say that the predicted values $\hat{Y_i}$ 


```python
pm.summary(kernel_trace)
```

```python
post_gammas = pm.summary(kernel_trace)['mean'].values[:-1]
y_pred = np.array([post_gammas[i]*kernel[i] 
                   for i in range(n_knots)]).sum(axis=0)
```

```python
idx = np.argsort(x)

with sns.plotting_context('talk'):
    plt.scatter(x[idx], y[idx], label='data')
    plt.scatter(x[idx], y_pred[idx], marker='x', label='prediction')
    _ = [plt.vlines(knot, -1, 1) for knot in knots]
    plt.legend()
```

So the mean response is pretty close most of the time, although we don't see a nice sine curve. 

We can produce a more informative prediction, by sampling from gamma posteriors (to get the mean of the response, $\mu$) and sampling from the posterior of epsilon (to get the standard deviation of the response ($\epsilon$) and then we can sample from *that* distribution to get a distribution of predictions.  

```python
ppc = pm.sample_ppc(kernel_trace, model=kernel_model, samples=100)
fig, ax = plt.subplots()

ax.plot(x, ppc['yl'].T, 'o', alpha=0.05, color=cols[1], label='prediction')
ax.plot(x, y, 'o', color=cols[0], label='data')
_ = ax.set_xlabel('$x$'); _ = ax.set_ylabel('$f(x)$')
```

This plot clearly demonstrates that each predicted value is actually a gaussian distribution of predictions  The width of the distribution is the unexplained variance.
