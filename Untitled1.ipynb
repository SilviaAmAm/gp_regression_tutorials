```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def gauss(x, mu=1, sd=1):
    return np.exp(-((x-mu)/sd)**2)
```

```python
x1 = np.linspace(-3,4,100)
x2 = np.linspace(-3,4,10)
y1 = gauss(x1)
y2 = gauss(x2)
plt.plot(x1, y1)
plt.plot(x2, y2)
```
